## Basic SVM Implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create a synthetic data points
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=1, n_redundant=0)
```

```python
X
```

```python
y
```

```python
# Convert X to DataFrame with named columns
df = pd.DataFrame(X, columns=["Feature 1", "Feature 2"])

# Create scatterplot
sns.scatterplot(x="Feature 1", y="Feature 2", hue=y, data=df)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='linear')
```

```python
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
rbf = SVC(kernel='rbf')
rbf.fit(X_train, y_train)
y_pred = rbf.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
polynamial = SVC(kernel='poly')
polynamial.fit(X_train, y_train)
y_pred = polynamial.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
sigmoid = SVC(kernel='sigmoid')
sigmoid.fit(X_train, y_train)
y_pred = sigmoid.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

## Hyperparametere Tuning With SVC

```python
from sklearn.model_selection import GridSearchCV
```

```python
 
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
grid=GridSearchCV(SVC(),param_grid=param_grid,refit=True,cv=5,verbose=3)
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
## Prediction
y_pred4=grid.predict(X_test)
print(classification_report(y_test,y_pred4))
print(confusion_matrix(y_test,y_pred4))
```
